# 🔬 MelanomaAI Training Monitor
Run cells top to bottom. Refreshes every 30 seconds automatically.

```python
import torch
print("GPU:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "CPU")
if torch.cuda.is_available():
    used = torch.cuda.memory_allocated(0)/1e9
    total = torch.cuda.get_device_properties(0).total_memory/1e9
    print(f"VRAM: {used:.1f}GB / {total:.1f}GB")
```

```python
import json, os, time, matplotlib.pyplot as plt
from IPython.display import clear_output

def show():
    clear_output(wait=True)
    plots = ["outputs/training_curves.png","outputs/confusion_matrix.png","outputs/roc_curve.png"]
    titles = ["Training Curves","Confusion Matrix","ROC Curve"]
    fig, axes = plt.subplots(1,3,figsize=(18,5),facecolor="#03070F")
    for ax,(p,t) in zip(axes,zip(plots,titles)):
        ax.set_facecolor("#0A101E")
        if os.path.exists(p):
            ax.imshow(plt.imread(p))
            ax.set_title(t,color="white")
        else:
            ax.text(0.5,0.5,f"{t}\n(training running...)",ha="center",va="center",color="#475569")
        ax.axis("off")
    plt.tight_layout()
    plt.show()
    for mf in ["outputs/metrics.json","outputs/metrics_phase1.json"]:
        if os.path.exists(mf):
            with open(mf) as f: m=json.load(f)
            phase = "FINAL" if "sensitivity" in m else "PHASE 1"
            print(f"\n--- {phase} RESULTS ---")
            print(f"Accuracy:    {m.get('accuracy',0)*100:.2f}%")
            print(f"AUC-ROC:     {m.get('auc_roc',0):.4f}")
            if "sensitivity" in m:
                print(f"Sensitivity: {m['sensitivity']*100:.2f}%")
                print(f"Specificity: {m['specificity']*100:.2f}%")
                print(f"F1-Score:    {m['f1_score']:.4f}")
            break
    else:
        print("\n⏳ Training not started. Run: python auto_train.py")

show()
```

```python
print("Auto-refreshing every 30s. Click stop button to cancel.")
while True:
    show()
    time.sleep(30)
```
